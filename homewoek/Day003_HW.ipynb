#### xmltodict

* XML content is value

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度

```python
import os, sys

# 打开文件
dirs = os.listdir( './data' )
```

```python
print("The number xml of dirs are {}.".format(len(dirs)))
```

```python
fh = open("./data/64_72hr_CH.xml", "r")
xml = fh.read()
fh.close()
```

```python
import xmltodict
d = dict(xmltodict.parse(xml))
```

```python
# 1. 請問高雄市有多少地區有溫度資料？
tmp = d['cwbopendata']['dataset']['locations']['location']
print("There are {} locations in 高雄市 to have temperate data.".format(len(tmp)))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
tmp[0]
```

```python
for idx, item in enumerate(tmp):
    print('Idx= {}, LocationName={}, '.format(idx, item['locationName']), end='')
    print('Time={}, '.format(item['weatherElement'][0]['time'][0]['dataTime']), end='')
    print('Temerate={}攝氏度'.format(item['weatherElement'][0]['time'][0]['elementValue']['value']))
    
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

loc_idx = 0
print('LocationName={}. '.format(tmp[loc_idx]['locationName']))
print('-'*50)
for idx, item in enumerate(tmp[loc_idx]['weatherElement'][0]['time']):
    try:
        print('Idx={}, '.format(idx), end='')
        print('Time={}, '.format(item['dataTime']), end='')
        print('Temerate={}攝氏度'.format(item['elementValue']['value']))
    except:
        print('-'*50)
        print('Err: ', item)
        print('-'*50)
```
